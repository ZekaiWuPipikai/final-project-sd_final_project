## Part 3: Data Preprocessing & Topic Modeling 

Part 3 will analyze that research question: **"What topics did people focus on while on Twitter in relation to the pandemic during the months of March and April 2020?"**

In part 3, we select a column called **"OriginalTweet"** from **"merged_df"** that contains the contents about what people post on Tweeter during the beginning of COVID-19 and do preprocessing procedure. Then, we have a list of cleaned text called **"Text_Cleaned"**. Subsequently, we utilize **"Text_Cleaned"** column to search for topics using LDA model. After inserting **count_vectorizer**, we have 5 different sets of topic with 9 different keywords in each topic set. And we also use a the t-SNE visualization tool to generate a graph with five different colors that represent each cluster of  topic set.

```python
!pip install nltk
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation as LDA
import re
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove RT (retweet sign)
    text = re.sub(r'rt[\s]+', '', text)
    # Remove mentions
    text = re.sub(r'@\S+', '', text)
    # Remove all non-alphabetic characters
    text = re.sub(r'\W', ' ', text)
    # Remove single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    # Remove leading and trailing whitespaces
    text = text.strip()
    # Remove stopwords and do stemming
    # text = " ".join([stemmer.stem(i) for i in text.split() if i not in stop_words])
    return text

merged_df['Text_Cleaned'] = merged_df['OriginalTweet'].apply(preprocess_text)
```

```python
merged_df[['Text_Cleaned','OriginalTweet']]
```

```python
documents = merged_df['Text_Cleaned']
```

```python
count_vectorizer = CountVectorizer(stop_words='english')
```

```python
count_data = count_vectorizer.fit_transform(documents)
```

```python
# Tweak the two parameters below (use int values below 15)
number_topics = 5
number_words = 9

# Create and fit the LDA model
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
```

```python
count_vectorizer.get_feature_names_out()
```

```python
words = list(count_vectorizer.get_feature_names_out())
```

```python
lda.components_.shape
```

```python
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names_out()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % (topic_idx+1))
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))

print_topics(lda, count_vectorizer, number_words)
```

```python
!pip install bokeh
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Transform the count data using LDA model
topic_weights = lda.transform(count_data)

# Apply t-SNE for dimensionality reduction
tsne_model = TSNE(n_components=2, random_state=42)
tsne_data = tsne_model.fit_transform(topic_weights)

# Calculate the dominant topic for each document
dominant_topic = topic_weights.argmax(axis=1)
merged_df['Dominant_Topic'] = dominant_topic

# Define topics and their corresponding colors
topics = ['Topic 1: COVID-19 impact on daily necessities',
          'Topic 2: Grocery shopping during COVID-19',
          'Topic 3: COVID-19 pandemic and consumer behavior',
          'Topic 4: COVID-19 impact on store workers and retail',
          'Topic 5: COVID-19 and food-related concerns']
topic_colors = ['red', 'green', 'blue', 'orange', 'purple']

# Plot the t-SNE visualization
plt.figure(figsize=(10, 8))
for i in range(number_topics):
    plt.scatter(tsne_data[:, 0][merged_df['Dominant_Topic'] == i],
                tsne_data[:, 1][merged_df['Dominant_Topic'] == i],
                color=topic_colors[i],
                alpha=0.7,
                label=topics[i])
plt.legend()
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('LDA Topic Visualization')
plt.show()
```

## Part 4: Sentiment Analysis

Part 4 will examine the research questions that **"What were people’s sentiments on Twitter in relation to the pandemic during the months of March and April 2020?"**

We import **“sentiment_textblob”** and **“sentiment_vader”** to analyze the texts from **“Text_Cleaned”** from **“merged_df.”** We will run these algorithms through every single tweet in the **“merged_df”** dataset and determine the sentiment of each tweet as well as the sentiment score. The results will be graphed through seaborn onto a bar plot and a line graph to represent the number of tweets that corresponds to a specific sentiment as well as the sentiment of tweets being tweeted over the course of March and April 2020.

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
import re
```

```python
def get_sentiment_textblob(text):
    sentiment = TextBlob(text).sentiment.polarity
    if sentiment > 0:
        return "Positive"
    elif sentiment < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
def get_sent_textblob_cont(text):
    return TextBlob(text).sentiment.polarity
```

```python
def get_sentiment_vader(text):
    sia = SentimentIntensityAnalyzer()
    sentiment = sia.polarity_scores(text)
    
    if sentiment['compound'] > 0.05:
        return "Positive"
    elif sentiment['compound'] < -0.05:
        return "Negative"
    else:
        return "Neutral"
```

```python
def get_sent_vader_cont(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)['compound']
```

```python
merged_df['sentiment_textblob'] = merged_df['Text_Cleaned'].apply(get_sentiment_textblob)
merged_df['sent_textblob_cont'] = merged_df['Text_Cleaned'].apply(get_sent_textblob_cont)
merged_df['sentiment_vader'] = merged_df['Text_Cleaned'].apply(get_sentiment_vader)
merged_df['sent_vader_cont'] = merged_df['Text_Cleaned'].apply(get_sent_vader_cont)
```

## Part 6: Emotional Analysis

Part 6 will analyze the research question that **"What were people’s emotions on Twitter in relation to the pandemic during the months of March and April 2020?"**

```python
#This function is used to install the NRCLex library, a tool that analyzes text for emotion content
!pip install nrclex
```

```python
#Import necessary packages 
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
import seaborn as sns
from nrclex import NRCLex
import matplotlib.dates as mdates
```

```python
# Function to get the top emotion from a text
def get_emotion_from_text(text):
    emotion = NRCLex(text)
    return emotion.top_emotions[0][0]

# Add emotion detection to the dataframe
merged_df['Emotion'] = merged_df['Text_Cleaned'].apply(get_emotion_from_text)

# Print the first few rows of the merged dataframe
merged_df
```

Essentially the function above is to add an "Emotions" section to the Dataset. 

```python
# Count the frequency of each unique value in the 'Emotion' column
emotion_counts = merged_df['Emotion'].value_counts()

# Create a bar plot for emotion frequency distribution
plt.figure(figsize=(10,6))
sns.barplot(x=emotion_counts.index, y=emotion_counts.values, alpha=0.8)
plt.title('Emotion Frequency Distribution')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Emotion', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

### Findings and Discussions for Bar Plot

The Emotion Frequency Distribution bar plot provides revealing insights into people's overall emotional response towards the pandemic as expressed in their tweets. Analyzing ten emotional categories using the "get_emotion_from_text" and nrclex functions, we find an unexpected surge in "positive" emotions. Among all emotions recorded, "positive" scores were strikingly high, with over 14,000 instances. This suggests that people maintained a generally positive outlook during March - April 2020. Considering this period marks the early stage of the pandemic, the result contradicts our initial hypothesis that people would exhibit more negative emotions.

The predominance of positive emotions during this time could be attributed to several factors. The mandate to "stay home" may have brought some relief to individuals who found the prospects of in-person activities daunting. Furthermore, the early stages of the pandemic were accompanied by a degree of ignorance regarding its impending severity and long-term impacts.

Besides the surge in positive emotions, other significant emotions were also prevalent. These include "fear" and "trust," each registering around 8,000 occurrences, and "negative" emotions, which accounted for approximately 5,500 instances. These figures, while less than the positive score, suggest that a substantial number of people experienced uncertainty and apprehension about what the future might hold in light of the pandemic.

In conclusion, our data shows that while there was an overwhelming sense of positivity expressed in tweets during the onset of the pandemic, likely due to the early stage naivety and the comfort of staying home, there was a rising undercurrent of fear, trust, and negativity. This undercurrent points to a growing concern about the potential future impacts of COVID-19.

```python
#Import necessary packages for visualization of emotions
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

#This line converts the 'TweetAt' column of the 'merged_df' DataFrame into datetime format using pandas
merged_df['TweetAt'] = pd.to_datetime(merged_df['TweetAt'])

# Filter data for March to April
filtered_df = merged_df[(merged_df['TweetAt'].dt.month >= 3) & (merged_df['TweetAt'].dt.month <= 4)]

emotion_over_time = filtered_df.groupby([filtered_df['TweetAt'].dt.date, 'Emotion']).size().reset_index(name='Counts')

emotion_over_time_pivot = emotion_over_time.pivot(index='TweetAt', columns='Emotion', values='Counts').fillna(0)

#This line initializes a new figure and axes object for the plot.
fig, ax = plt.subplots()
emotion_over_time_pivot.plot(kind='line', stacked=False, ax=ax)

#This line creates a line graph using the pivoted DataFrame. 
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

#This line rotates the x-axis labels by 45 degrees and right-aligns them for better readability.
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

#This line is used to adjust the visualization and display the results
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Emotion Distribution Over Time (March-April)')
plt.tight_layout()
plt.show()
```

### Discussion and Findings from the Emotions Frequency Distribution

Contrary to our initial hypothesis, which proposed an overwhelmingly negative emotional response towards the pandemic, our temporal analysis of tweet frequencies from March to April 2020 revealed a different trend. Rather than a prevailing negativity, we observed a surge of positivity during the early stages of the pandemic. This optimism, however, began to diminish over time, giving way to rising feelings of fear, mistrust, and negativity. This emotional shift was not sudden but happened gradually, mirroring the unfolding realities and growing uncertainties of the pandemic. Initially, positive emotions, possibly sparked by the prospect of staying home and the relative lack of knowledge about the pandemic's future impacts, dominated the discourse. However, as the pandemic's complexities began to be realized, our data showed a marked increase in negative emotions. As such, the hypothesis of a predominantly negative emotional response does not consistently align with our findings for the entire period. For further studies, we hypotesize that the negative emotions only became more prevalent later on, as the true implications of the pandemic became apparent and the initial wave of optimism receded in the face of mounting concerns and fears.

### Limitations about the function

While our findings provide valuable insights into public sentiment during the early stages of the pandemic, it's important to acknowledge the limitations inherent to our methodology. The primary constraint is that our analysis is confined to tweets from March to April 2020. While this period marks a significant phase at the onset of the pandemic, it does not capture the full evolution of emotions as the pandemic unfolded over subsequent months and years. As such, our data offers only a snapshot of the emotional landscape during this time and may not accurately represent the broader sentiment throughout the entire pandemic period.

Additionally, the "get_emotion_from_text" and nrclex functions, which we employed for emotion detection, categorize emotions broadly as "positive," without delving into the nuanced specifics of what these positive emotions entail—happiness, satisfaction, interest, etc. Therefore, our conclusion about the predominance of "positive" sentiments during the initial stages of the pandemic encompasses a wide range of specific feelings. Future studies could address these limitations by extending the timeline of data collection to include subsequent phases of the pandemic, and by refining the emotion detection methods to distinguish between discrete positive emotions, thereby painting a more accurate picture of public sentiment throughout the pandemic.
